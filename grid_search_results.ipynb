```python
def calc_file_rmse(filename):
    df = pd.read_csv(f'results/grid_search/{filename}')
    df['config'] = df['features_elimination_prc'].astype(str)+"_"+df['epsilon_factor'].astype(str)+"_"+df['dist_functions'].astype(str)
    df_max_vals = df.groupby(['feature_prc']).agg({'accuracy':'max'}).reset_index().sort_values(by='feature_prc')
    configs_arr = df['config'].unique()
    rmse_list = [
    (conf, math.sqrt(sklearn.metrics.mean_squared_error(df_max_vals.accuracy, df.sort_values(by='feature_prc').loc[df['config'] == conf, 'accuracy']))  )
    for conf in configs_arr
    ]
    rmse_df = pd.DataFrame(rmse_list, columns=['config', 'rmse'])
    return {filename: ({'rmse_df': rmse_df}, {'df_max_vals': df_max_vals})}

dfs = [calc_file_rmse(filename) for filename in filenames]    
# rmse_data = pd.concat(dfs)    
```
