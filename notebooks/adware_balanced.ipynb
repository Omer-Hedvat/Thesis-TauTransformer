```python
import numpy as np
import os
import pandas as pd
from main import read_from_csv
```

```python
df = read_from_csv('data/original_datasets/adware.csv', {'label_column': 'Class'})
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
df.head()
```

```python
new_cols = [col.strip().lower().replace(' ', '_') for col in df.columns]
old_cols = df.columns
headers_dict = {key: value for key, value in zip(old_cols, new_cols)}

df.rename(headers_dict, axis=1, inplace=True)
df.head()
```

```python
df['label'].value_counts()
```

```python
rows = df['label'].value_counts().min()
```

```python
df.dtypes
```

```python
cols_to_drop = ['flow_id', 'source_ip', 'source_port', 'destination_ip', 'destination_port', 'protocol', 'timestamp']
df.drop(labels=cols_to_drop, axis=1, inplace=True)
df.head()
```

```python
df_balanced = pd.concat([df.loc[df['label']==0].head(rows), df.loc[df['label']==1].head(rows), df.loc[df['label']==2].head(rows), df.loc[df['label']==3].head(rows), df.loc[df['label']==4].head(rows), df.loc[df['label']==5].head(rows)])
```

```python
df_balanced.label.value_counts()
```
