```python
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv("all_datasets_results.csv")
dataset = 'glass'
```

```python
data = data[data['dataset'] == dataset]
```

```python
data
```

```python
A_type = ['random_features','fisher','relief','Chi_square']
B_type = data.iloc[:,8:].columns
```

```python
df = pd.DataFrame(data={'dataset': [dataset]})
```

```python
for a in A_type:
    for b in B_type:
        df[a+' '+b] = stats.ttest_ind(data[a],data[b], alternative ='less')[1]
```
